```python

UMBRAL = 0.95      
static_cols = [
    'PLAZA_CVE',
       'NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES', 'MTS2VENTAS_NUM',
       'PUERTASREFRIG_NUM', 'CAJONESESTACIONAMIENTO_NUM', 'LATITUD_NUM',
       'LONGITUD_NUM', 'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA',
       'DATASET'
   
]


tienda_static = (
    venta_vs_meta
    .groupby('TIENDA_ID')[static_cols]
    .first()               
)


resumen = (
    venta_vs_meta
    .groupby('TIENDA_ID')['meta_alcanzada']
    .agg(total_meses='size', meses_cumplidos='sum')
)

resumen['proporcion'] = resumen['meses_cumplidos'] / resumen['total_meses']
resumen['exito_70']   = (resumen['proporcion'] >= UMBRAL).astype(int)

dataset_tienda = tienda_static.join(resumen[['exito_70']])

dataset_tienda.sample(5)
```
